## Housing Case Study(using Decision Tree)

#### Problem Statement:

Consider a real estate company that has a dataset containing the prices of properties in the Delhi region. It wishes to use the data to optimise the sale prices of the properties based on important factors such as area, bedrooms, parking, etc.

Essentially, the company wants â€”


- To identify the variables affecting house prices, e.g. area, number of rooms, bathrooms, etc.

- To create a linear model that quantitatively relates house prices with variables such as number of rooms, area, number of bathrooms, etc.

- To know the accuracy of the model, i.e. how well these variables can predict house prices.

# Preparation the Data for Modelling

##### Binary Encoding

```python
varlist=['mainroad','guestroom', 'basement', 'hotwaterheating', 'airconditioning','prefarea']

housing[varlist]=housing[varlist].apply(lambda x:x.map({ "no":0,"yes":1}))
```

```python
housing.head()
```

##### Dummy encoding

```python
df_dummy=pd.get_dummies(housing["furnishingstatus"],drop_first=True)
housing=pd.concat([housing,df_dummy],axis=1)

housing.drop(columns=["furnishingstatus"],inplace=True)
```

### train-test split

```python
df_train,df_test=train_test_split(housing,train_size=0.70,random_state=100)
```

```python
df_train.head()
```

#### Re-Scaling

```python
# instantiate an object
scaler=MinMaxScaler()
```

```python
# fit on data
num_vars=['price', 'area', 'bedrooms', 'bathrooms', 'stories','parking']

df_train[num_vars]=scaler.fit_transform(df_train[num_vars])
```

```python
df_train[num_vars]
```

# Training the Model

```python
plt.figure(figsize=(12,10))
sns.heatmap(df_train.corr(),annot=True,cmap="Blues")
plt.show()
```

```python
df_train.head()
```

```python
# Dividing into X_train,y_train

X_train=df_train.drop("price",axis=1)
y_train=df_train["price"]
```

# Forward Selection Technique

```python
# add a constant
X_train_sm=sm.add_constant(X_train["area"])

# create first model and fit
lr_1=sm.OLS(y_train,X_train_sm).fit()

# params
lr_1.params

# summary
lr_1.summary()
```

```python
# add another variable bathrooms
X_train_sm=sm.add_constant(X_train[["area","bathrooms"]])

# create another model and fit
lr_2=sm.OLS(y_train,X_train_sm).fit()

# params
lr_2.params

# summary
lr_2.summary()
```

```python
# add another variable bedrooms
X_train_sm=sm.add_constant(X_train[["area","bathrooms","bedrooms"]])

# create another model and fit
lr_3=sm.OLS(y_train,X_train_sm).fit()

# params
lr_3.params

# summary
lr_3.summary()
```

# Backward Selection Technique

### Adding all the variables to the model

```python
# build a model with all the variables
X_train_sm=sm.add_constant(X_train)

# create model and fit
lr_1=sm.OLS(y_train,X_train_sm).fit()

#params
lr_1.params

#summary
lr_1.summary()
```

# Checking VIFs

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# Create a dataframe that will contain the names of all the feature variables and their respective VIFs
vif = pd.DataFrame()
vif['Features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

We could have:
- High p-value,high VIF
- High-low : 
    - High p,low VIF  ----------> remove these first
    - Low p,High VIF  ----------> remove these after the ones above
- Low p-value,low VIF

# Using Decision Trees

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import plot_tree
```

##### train-test split

```python
df_train,df_test=train_test_split(housing,train_size=0.7,random_state=100)
```

#### rescaling

```python
scaler=MinMaxScaler()

df_train["price"]=scaler.fit_transform(df_train[["price"]])
df_test["price"]=scaler.transform(df_test[["price"]])
```

#### Dividinig into X_train,y_train,X_test,y_test

```python
X_train=df_train.drop(columns=["price"],axis=1)
y_train=df_train["price"]
X_test=df_test.drop(columns=["price"],axis=1)
y_test=df_test["price"]
```

```python
X_test.shape
```

## Fitting the Model

```python
dt=DecisionTreeRegressor(max_depth=4,min_samples_leaf=10,random_state=42)

dt.fit(X_train,y_train)
```

## Model interpretation

```python
plt.figure(figsize=(90,50))
plot_tree(dt,feature_names=list(X_train.columns),filled=True)
plt.show()
```

## Model evaluation.

```python
y_train_pred=dt.predict(X_train)
y_test_pred=dt.predict(X_test)
print("Train set performance : " )
print(r2_score(y_train,y_train_pred))
print("-"*50)
print("Test set performance : " )
print(r2_score(y_test,y_test_pred))
```

We can definitely perform more hyperparameter tuning here and improve the performance of the models
